```python
titanic_url = "https://raw.githubusercontent.com/pcsanwald/kaggle-titanic/master/train.csv"
titanic = pd.read_csv(titanic_url)
titanic.head()
```

```python
titanic.shape
```

```python
json_string = titanic.to_json(orient="records")

import json
json_obj = json.loads(json_string)
print(type(json_obj))
```

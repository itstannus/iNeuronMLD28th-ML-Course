```python
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv"
crime = pd.read_csv(url)
crime.head()
```

```python
#Convert the type of the column Year to datetime64
```

```python
crime.dtypes
```

```python
# crime['Year'].astype('datetime64[s]').head()
```

```python
crime.Year = pd.to_datetime(crime['Year'] ,infer_datetime_format=True, format='%Y')
crime.dtypes
```

```python
crime.head()
```

```python
#Set the Year column as the index of the dataframe
```

```python
crime.set_index('Year', inplace=True)
```

```python
crime.loc['1960-01-01']
```

```python
#Group the year by decades and sum the values
```

```python
# Uses resample to sum each decade i.e. sum for all columns and max value only in Popuation column as population is cumulative so consider only max
crimes = crime.resample('10AS').sum() # AS, YS year start frequency
```

```python
crimes
```

```python
# Uses resample to get the  max value only for the "Population" column
population = crime['Population'].resample('10AS').max()
population
```

```python
# Updating the "Population" column
crimes['Population'] = population
crimes
```
